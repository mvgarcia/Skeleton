```python
plt.figure(figsize=(12,4))
ax1=plt.subplot(131)
for i in bsk1:
    i=np.array(i)
    xy=np.array([abacus1[int(i[0])][0:2],abacus1[int(i[1])][0:2]])
    plt.plot(xy[:,0],xy[:,1],color="k",linewidth=0.50,label="$\\beta=1$")
plt.setp(ax1.get_yticklabels(), visible=True)
plt.ylabel("$y\;(Mpc)$")
plt.legend(("$\\beta=1$",),loc='upper left')
plt.axis("image")

ax2=plt.subplot(132)
for i in bsk2:
    i=np.array(i)
    xy=np.array([abacus1[int(i[0])][0:2],abacus1[int(i[1])][0:2]])
    plt.plot(xy[:,0],xy[:,1],color="k",linewidth=0.50)
plt.xlabel("$x\;(Mpc)$")
plt.setp(ax2.get_yticklabels(), visible=False)
plt.legend(("$\\beta=2$",),loc='upper left')
plt.axis("image")

ax3=plt.subplot(133,sharey=ax2)
for i in bsk5:
    i=np.array(i)
    xy=np.array([abacus1[int(i[0])][0:2],abacus1[int(i[1])][0:2]])
    plt.plot(xy[:,0],xy[:,1],color="k",linewidth=0.50)
plt.setp(ax3.get_yticklabels(), visible=False)
plt.legend(("$\\beta=5$",),loc='upper left')
plt.axis("image")


plt.savefig("beta125_2.pdf")
```
